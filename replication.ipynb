### 1. Loading data and preprocessing

```python
import os
import numpy as np
import pandas as pd
import datetime as dt
from scipy.stats import entropy

# specify the location of the framing data
data_dir = '/Users/dcard/Dropbox/CMU/ARK/compuframes/Analysis/Amber/data/'
data_file = os.path.join(data_dir, 'corrected_for_stata.csv')

# read the data into a pandas data frame
data = pd.read_csv(data_file, header=0, index_col=0)

# exclude articles marked as "irrelevant" and those from before 1980
data = data.loc[(data['Irrelevant']==0) & (data['Year'] >= 1980)]

# combine year, month, day into a single date variable
data['date'] = data.apply(lambda row: pd.Timestamp(dt.date(row['Year'], row['Month'], row['Day'])), axis=1)

# determine the quarter (1/4 year) from the date
data['quarter'] = data.apply(lambda row: row['date'].quarter, axis=1)

# list the columns that contain framing variables
# each of these columns contains the estimated probability that each article makes use of the corresponding frame
frames = ['Capacity_and_resources',
'Crime_and_punishment',
'Cultural_identity',
'Economic',
'External_regulation',
'Fairness_and_equality',
'Health_and_safety',
'Legality_jurisdiction',
'Morality',
'Policy_prescription',
'Political',
'Public_sentiment',
'Quality_of_life',
'Security_and_defense',
'Other']

# compute the net tone
data['tone'] = data['Pro'] - data['Anti']

# list the columns that contain tone variables
# each of the first three of these contains the estimated probability that each article is pro, neutral, or anti, respectively
tone_cols = ['Pro', 'Neutral', 'Anti', 'tone']

# create a new column for the sum of the framing columns
data['all_frames_story'] = 0

# add in the contribution of each column
for f in frames:
    data['all_frames_story'] += data[f]
    
# display a summary
print data['all_frames_story'].describe()
```

```python
# create a dummy variable = 1 for all articles
data['stories'] = 1
    
# create a grouping of articles by year/quarter
groups = data.groupby(['Year', 'quarter'])

# create a new dataframe "grouped", which is what we will work with
grouped = pd.DataFrame()

# add up the total number of articles per quarter and store in grouped
grouped['stories'] = groups.aggregate(np.sum)['stories']

# For each framing/tone column, take the mean of the articles in each group and store in grouped.
# Each of these columns will contain the average across all articles in the quarter of 
# the estimated probability of each article making use of the corresponding frame
for f in frames + tone_cols:
    grouped[f] = groups.aggregate(np.mean)[f]

# take the sum of all mean probabilities in each quarter
grouped['all_frames'] = 0
for f in frames:
    grouped['all_frames'] += grouped[f]

# for each framing dimension divide the mean probability by the sum of the mean probabilities in the quarter
for f in frames:
    grouped['p_' + f] = grouped[f] / grouped['all_frames']

# replicating Amber's explicit calculations, not needed    
#grouped['entropy.bruteforce'] = 0
#for f in frames:
#    grouped['lnp_' + f] = np.log(grouped['p_' + f])
#    grouped['p_lnp_' + f] = grouped['p_' + f] * grouped['lnp_' + f]
#    grouped['entropy.bruteforce'] -= grouped['p_lnp_' + f]

# compute the entropy of the proportional mean probabilties calculated above
grouped['entropy'] = grouped.apply(lambda row: entropy([row['p_' + f] for f in frames]), axis=1)

# print a summary of the entropy variable
print grouped['entropy'].describe()

# compute the correlation between the number of articles per quarter and the entropy of framing
print grouped[['stories', 'entropy']].corr()
```

```python
# specify the location of the mood data file
data_file = os.path.join(data_dir, 'quarter_smoothed.csv')

# read the mood data into a dataframe
mood_by_quarter = pd.read_csv(data_file, header=0, index_col=0)

# take the mood from the last quarter of 1979, since we're going to need it later
mood_1979_4 = mood_by_quarter[mood_by_quarter['quarter'] == 4].loc[1979]['mood']

# use the same range of dates as the framing data (exclude before 1980 and after 2012)
mood_by_quarter = mood_by_quarter.loc[(mood_by_quarter.index >= 1980) & (mood_by_quarter.index < 2013)]

# convert each quarter to a group (year/quarter)
groups = mood_by_quarter.groupby([mood_by_quarter.index,  'quarter'])

# create a new dataframe
mood_grouped = pd.DataFrame()

# store the mood from each quarter using the same index as the framing data
mood_grouped['mood'] = groups.aggregate(np.sum)['mood']

# copy the mood data into the grouped dataframe (with the framing data)
grouped['mood'] = mood_grouped['mood']

# store the vector of mood values from the previous quarter as a separate vector
prev_mood = grouped['mood'].shift().as_matrix()

# set the first value to the mood from the last quarter of 1979
prev_mood[0] = mood_1979_4

# save the mood from the previous quarter as a new column in "grouped"
grouped['prev_mood'] =  prev_mood
```

```python
# create a "dominance" variable that is the negative entropy
grouped['dominance'] = - grouped['entropy']

# create a "salient" variable that is 1 if the number of stories per quarter >= 350, otherwise 0
grouped['salient'] = grouped.apply(lambda row: int(row['stories'] >= 350), axis=1)

# create an event variable; set those quarters with major events to 1
grouped['event'] = 0
grouped['event'].loc[1990,4] = 1.0
grouped['event'].loc[1994,4] = 1.0
grouped['event'].loc[2000,2] = 1.0
grouped['event'].loc[2001,4] = 1.0
grouped['event'].loc[2006,2] = 1.0

# add in interactions
# between tone and number of stories per quarter
grouped['toneXstories'] = grouped['tone'] * grouped['stories']
# between tone and the "salient" variable
grouped['toneXsalient'] = grouped['tone'] * grouped['salient']
# between tone and the "entropy" variable
grouped['toneXentropy'] = grouped['tone'] * grouped['entropy']
# between tone and the presence of an event in a quarter
grouped['toneXevent'] = grouped['tone'] * grouped['event']
```

### 3. Regressions:

```python
# Regression 1: run an unregularized linear regression against the set of variable below
pd.ols(y=grouped['mood'], x=grouped[['prev_mood', 'tone', 'stories', 'toneXstories', 'entropy', 'toneXentropy']])
```

```python
# verify and plot results with scikit-learn

from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
%matplotlib inline


# use the same predictors as above
columns = ['prev_mood', 'tone', 'stories', 'toneXstories', 'entropy', 'toneXentropy']

X = grouped[columns].as_matrix()

# use the mood as the variable to predicted
y = grouped['mood'].as_matrix()

# create a basic linear regression model
ols = LinearRegression()

# fit the model
ols.fit(X=X, y=y)

# get the model's predictions
pred = ols.predict(X)

# display and plot the results
print "Intercept:", ols.intercept_
for index, c in enumerate(columns):
    print c + ":", ols.coef_[index]
plt.figure(figsize=(8,6))
x_range = np.arange(len(y))
plt.plot(x_range, y, 'k--', linewidth=2, label='Mood')
plt.plot(x_range, pred, 'r-', label='Predicted')
plt.legend(loc='upper left')
plt.show()
```

### 4. My own quick test with Lasso

```python
# Try running some quick Lasso regressions using the same set of variables to see what survives

from sklearn.linear_model import Lasso, Ridge, LinearRegression
from sklearn.preprocessing import scale
import matplotlib.pyplot as plt
%matplotlib inline

# select the same columns as Regression 1
cols_to_scale = ['prev_mood', 'toneXstories', 'entropy', 'stories', 'tone', 'toneXentropy']

# scale the predictors to mean = 0, sd = 1
scaled = scale(grouped[cols_to_scale])

# take the scaled values as the design matrix
X = scaled

# take mood (unchanged) as the values to be predicted
y = grouped['mood'].as_matrix()

# loop through various values of regularization
for i in range(1,-3,-1):
    # set the regularization strength
    alpha = np.power(np.sqrt(10.0),i)
    # create a Lasso model
    model = Lasso(normalize=False, alpha=alpha)
    # fit the model
    model.fit(X=X, y=y)
    # take the model predictions
    pred = model.predict(X)
    # display the results and plot
    print "Regluarization strength =", alpha
    print "================================="
    print "Intercept:", model.intercept_
    for index, c in enumerate(cols_to_scale):
        if model.coef_[index] != 0:
            print c + ":", model.coef_[index]
    plt.figure(figsize=(8,6))
    x_range = np.arange(len(y))
    plt.plot(x_range, y, 'k--', linewidth=2, label='Mood')
    plt.plot(x_range, pred, label='Predicted')
    plt.legend(loc='upper left')
    plt.show()
```
